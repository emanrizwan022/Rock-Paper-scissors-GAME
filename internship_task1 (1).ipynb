```python
import random
import ipywidgets as widgets
from IPython.display import display

# Function to get computer's choice
def get_computer_choice():
    choices = ["rock", "paper", "scissors"]
    return random.choice(choices)

# Function to determine the winner
def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "It's a tie!"
    elif (user_choice == "rock" and computer_choice == "scissors") or \
         (user_choice == "scissors" and computer_choice == "paper") or \
         (user_choice == "paper" and computer_choice == "rock"):
        return "You win!"
    else:
        return "Computer wins!"

# Function to handle the game logic
def play_game(b):
    user_choice = choice_dropdown.value
    if user_choice not in ["rock", "paper", "scissors"]:
        result_label.value = "Invalid choice. Please select rock, paper, or scissors."
    else:
        computer_choice = get_computer_choice()
        result = determine_winner(user_choice, computer_choice)
        result_label.value = f"Computer chose: {computer_choice}\n{result}"
        play_again_button.visible = True  # Show the "Play Again" button

# Function to handle the "Play Again" button click
def play_again(b):
    result_label.value = ""  # Clear the result label
    choice_dropdown.value = "rock"  # Reset the dropdown to the default value
    play_again_button.visible = False  # Hide the "Play Again" button

# Create widgets
instructions = widgets.HTML(value="<h2>Rock-Paper-Scissors Game</h2><p>Select your choice from the dropdown and click 'Submit' to see the result.</p>")

choice_dropdown = widgets.Dropdown(
    options=["rock", "paper", "scissors"],
    value="rock",
    description="Your choice:",
    style={'description_width': 'initial'}
)

submit_button = widgets.Button(
    description="Submit",
    button_style='primary'
)
submit_button.on_click(play_game)

result_label = widgets.Label(value="")

play_again_button = widgets.Button(
    description="Play Again",
    button_style='success',
    visible=False  # Initially hidden
)
play_again_button.on_click(play_again)

# Display widgets
display(instructions, choice_dropdown, submit_button, result_label, play_again_button)
```
